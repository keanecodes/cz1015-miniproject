```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
athlete = pd.read_csv('./athlete_events.csv')
athlete.head()
```

```python
noc = pd.read_csv('./noc_regions.csv')
noc.columns = ['NOC', 'Region', 'Notes']
noc.head()
```

```python
mergedData = pd.merge(athlete, noc, on='NOC', how='left')
mergedData.head()
```

```python
print(mergedData.shape)
print(len(mergedData['Name'].unique()))
```

```python
#print(mergedData["Region"].value_counts())
sb.catplot(y = "Region", data = mergedData, kind = "count", height=50)
```
